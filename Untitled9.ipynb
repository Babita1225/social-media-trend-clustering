```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
import re
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+|#", "", text)
    text = re.sub(r"[^A-Za-z\s]", "", text)
    text = text.lower()
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text

df = pd.read_csv("training.1600000.processed.noemoticon.csv", encoding='latin-1', header=None, delimiter=',')
df.columns = ["target", "id", "date", "flag", "user", "text"]
df['clean_text'] = df['text'].astype(str).apply(clean_text)

# Use a small sample for testing
sample = df.sample(5000, random_state=42)
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(sample['clean_text'])
# Try different K values
scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    scores.append(silhouette_score(X, kmeans.labels_))

# Plot to choose best K
plt.plot(range(2, 10), scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Choosing K using Silhouette Score')
plt.show()
# Use chosen K (e.g., K=4)
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

# Evaluate
sil_score = silhouette_score(X, labels)
db_index = davies_bouldin_score(X.toarray(), labels)

print("Silhouette Score:", sil_score)
print("Davies–Bouldin Index:", db_index)
dbscan = DBSCAN(eps=0.5, min_samples=5, metric='cosine')
db_labels = dbscan.fit_predict(X)

# Filter out noise points (-1)
filtered = db_labels != -1
if sum(filtered) > 0:
  print("Silhouette Score (DBSCAN):", silhouette_score(X[filtered], db_labels[filtered]))
```
